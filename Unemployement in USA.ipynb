```python
import pandas as pd
unrate = pd.read_csv("unrate.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
x_values = unrate["DATE"][0:12]
y_values = unrate["UNRATE"][0:12]
plt.plot(x_values , y_values)
plt.show()
```

## FINDINGS

It's an data visualization project about unemployement in USA.Basic using of matplotlib library and pyplot module.

```python
x_values = unrate["DATE"][0:12]
y_values = unrate["UNRATE"][0:12]
plt.plot(x_values , y_values)
plt.xticks(rotation=90)
plt.show()
```

```python
x_values = unrate["DATE"][0:12]
y_values = unrate["UNRATE"][0:12]
plt.plot(x_values , y_values)
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

## final
We found monthly unemployment trends at 1948. We can research rate differences between months.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['UNRATE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['UNRATE'])
ax2.set_title('Monthly Unemployment Rate, 1949')

plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['UNRATE'])
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['UNRATE'])
ax3.plot(unrate[24:36]['DATE'], unrate[24:36]['UNRATE'])
ax4.plot(unrate[36:48]['DATE'], unrate[36:48]['UNRATE'])
ax5.plot(unrate[48:60]['DATE'], unrate[48:60]['UNRATE'])
plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"],c="red")
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"],c="blue")
plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(10,6))
plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"],c="red")
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"],c="blue")
plt.plot(unrate[24:36]["MONTH"],unrate[24:36]["UNRATE"],c="green")
plt.plot(unrate[36:48]["MONTH"],unrate[36:48]["UNRATE"],c="orange")
plt.plot(unrate[48:60]["MONTH"],unrate[48:60]["UNRATE"],c="black")
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black'] 
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i],label = str(1948 + i))
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.title('Monthly Unemployment Trends, 1948-1952')
plt.legend(loc="upper left")
plt.show()
```
